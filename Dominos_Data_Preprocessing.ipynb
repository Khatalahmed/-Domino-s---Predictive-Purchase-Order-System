```python
def parse_dates(date):
    for fmt in ('%d-%m-%Y', '%d/%m/%Y', '%m/%d/%Y'):
        try:
            return pd.to_datetime(date, format=fmt)
        except ValueError:
            continue
    return pd.NaT
df_sales['order_date'] = df_sales['order_date'].apply(parse_dates)  
df_sales['quantity'] = pd.to_numeric(df_sales['quantity'], errors='coerce')
```

```python
merged_df.drop(columns=['pizza_name_y', 'pizza_ingredients_y'], inplace=True)
merged_df.rename(columns={
    'pizza_name_x': 'pizza_name',
    'pizza_ingredients_x': 'pizza_ingredients'
}, inplace=True)
```

```python
# handling duplicates values
merged_df.duplicated().sum()
merged_df = merged_df.drop_duplicates()
```

```python
# Extracting the day of the week
merged_df['day_of_week'] = merged_df['order_date'].dt.day_name()
merged_df['day_of_week_num'] = merged_df['order_date'].dt.weekday

# Extracting the month
merged_df['month'] = merged_df['order_date'].dt.month # month_name()
merged_df['month_name'] = merged_df['order_date'].dt.month_name() # month_name()

# Createing a new column for the week number
merged_df['week'] = merged_df['order_date'].dt.isocalendar().week
# Extracting the year
merged_df['year'] = merged_df['order_date'].dt.year
```

### Applying holidays to the data

```python
import holidays

# using US holidays
us_holidays = holidays.US()

# Creating a 'holiday' column that indicates if the order_date was a holiday
merged_df['holiday'] = merged_df['order_date'].apply(lambda x: 1 if x in us_holidays else 0)

# Checking if the holiday flag works
merged_df[['order_date', 'holiday','day_of_week']].head()
```

## Adding weekend promotions

```python
# Creating a 'promotional_period' flag for weekends 
merged_df['promotion'] = merged_df['order_date'].dt.weekday.apply(lambda x: 1 if x >= 5 else 0)

# Checking if the promotion flag works
merged_df[['order_date', 'promotion']].head()
```

### daily,weekly and monthly pizza sales quantity Visualization

```python
# Ensure 'order_date' is a datetime type
merged_df['order_date'] = pd.to_datetime(merged_df['order_date'])

# Set 'order_date' as the index
merged_df.set_index('order_date', inplace=True)

# Resample to calculate daily sales
daily_sales = merged_df['unit_price'].resample('D').sum()
daily_quantity = merged_df['quantity'].resample('D').sum()


# Display the daily sales
daily_sales
```

```python
daily_quantity
```

```python
import matplotlib.pyplot as plt
# Plot both on the same graph
plt.figure(figsize=(10, 6))
plt.plot(daily_sales, label='Daily Sales', color='blue')
plt.plot(daily_quantity, label='Daily Quantity', color='orange')
plt.title('Daily Sales and Quantity')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python

# Resample to calculate week sales
week_sales = merged_df['unit_price'].resample('W').sum()
week_quantity = merged_df['quantity'].resample('W').sum()


# Display the week sales
week_sales
```

```python
plt.figure(figsize=(10, 6))
plt.plot(week_sales, label='Weekly Sales', color='blue')
plt.plot(week_quantity, label='Weekly Quantity', color='orange')
plt.title('Weekly Sales and Quantity')
plt.xlabel('Date')
plt.ylabel('Values')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Resample to calculate week sales
Month_sales = merged_df['unit_price'].resample('M').sum()
Month_quantity = merged_df['quantity'].resample('M').sum()

9999
# Display the week sales
len(Month_sales)
```

### Top 15 selling pizza

```python
import matplotlib.pyplot as plt

# Group by pizza_name and calculate total sales
pizza_sales = merged_df.groupby('pizza_name')['total_price'].sum()

# Sort sales in descending order and select the top 15
top_15_pizzas = pizza_sales.sort_values(ascending=False).head(15)

# Plot the top 15 selling pizzas
palette = sns.color_palette("coolwarm", n_colors=15)
plt.figure(figsize=(12, 8))
top_15_pizzas.plot(kind='bar', color=palette, edgecolor='black')
plt.title('Top 15 Selling Pizzas', fontsize=16)
plt.xlabel('Pizza Name', fontsize=14)
plt.ylabel('Total Sales', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

### pizza size - sales

```python
# Ploting pizza size distribution
size_sales = merged_df.groupby('pizza_size')['quantity'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=size_sales, x='pizza_size', y='quantity', palette='rocket')
plt.title('Sales by Pizza Size')
plt.xlabel('Pizza Size')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Ploting pizza category distribution
category_sales = merged_df.groupby('pizza_category')['quantity'].sum().reset_index()

plt.figure(figsize=(8,5))
sns.barplot(data=category_sales, x='pizza_category', y='quantity', palette='coolwarm')
plt.title('Sales by Pizza Category')
plt.xlabel('Pizza Category')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
# Aggregate sales by day of the week
sales_by_day = merged_df.groupby('day_of_week')['total_price'].sum().reset_index()

# Plot sales by day of the week
plt.figure(figsize=(10,5))
sns.barplot(data=sales_by_day, x='day_of_week', y='total_price', palette='viridis')
plt.title('Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Aggregate sales by month
sales_by_day = merged_df.groupby('month_name')['total_price'].sum().reset_index()

# Plot sales by day of the week
plt.figure(figsize=(14,6))
sns.barplot(data=sales_by_day, x='month_name', y='total_price', palette='viridis')
plt.title('Sales by month')
plt.xlabel('name of the month')
plt.ylabel('Total Sales')
plt.show()
```
